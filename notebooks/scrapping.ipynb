```python
from datetime import datetime
import quandl
import sys
import os
from secedgar.filings import Filing, FilingType
import xml.etree.ElementTree as ET
import pandas as pd
```

```python
filing = Filing(cik_lookup='aapl',
                filing_type=FilingType.FILING_4,
                start_date=datetime(2019, 7, 1),
                end_date=datetime(2020, 6, 30))
```

```python
filing.save('~/path/to/dir')
```

```python
files_2 = os.listdir('~/path/to/dir/aapl/4')
```

```python
files = files_2[:4] + files_2[5:]
```

```python
data = pd.DataFrame(columns= ['transactiondate', 'transactionshares', 'transactionpricepershare',
                             'transactioncode', 'officerTitle', 'rptOwnerName', 'issuername', 'ticker'])
```

```python
for file in files:
    form_4 = open('~/path/to/dir/aapl/4/' + file, "r")
    print(file)
    sec_form_raw = form_4.read()
    sec_form_processed = sec_form_raw
    index = sec_form_processed.find("</SEC-HEADER>")
    sec_form_processed = sec_form_processed[:index] + "</ACCEPTANCE-DATETIME>\n" + sec_form_processed[index:]
    index = sec_form_processed.find('<?xml version="1.0"?>')
    sec_form_processed = sec_form_processed[:index] + sec_form_processed[(index+len('<?xml version="1.0"?>')):]
    index = sec_form_processed.find("</DOCUMENT>")
    sec_form_processed = sec_form_processed[:index] + "</DESCRIPTION>\n</FILENAME>\n</SEQUENCE>\n</TYPE>\n"  + sec_form_processed[index:]
    
    root = ET.fromstring(sec_form_processed)
    row = {}
    for x in root.iter('transactionDate'):
        row['transactiondate'] = x[0].text
    for x in root.iter('transactionShares'):
        row['transactionshares'] = x[0].text 
    for x in root.iter('transactionPricePerShare'):
        row['transactionpricepershare'] = x[0].text
    for x in root.iter('transactionCode'):
        row['transactioncode'] = x.text
    for x in root.iter('officerTitle'):
        row['officerTitle'] = x.text
    for x in root.iter('rptOwnerName'):
        row['rptOwnerName'] = x.text
    for x in root.iter('issuerName'):
        row['issuername'] = x.text
    for x in root.iter('issuerTradingSymbol'):
        row['ticker'] = x.text

    data = data.append(row, ignore_index=True)



```

```python
data.to_csv('apple_last_year.csv')
```
